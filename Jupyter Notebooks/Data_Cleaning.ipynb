```python
df = df.drop_duplicates()
df.duplicated().sum()
```

Dealing with Cholesterol Column

```python
## cholesterol min value is zero , whyy (while mean is 210) ?? How manyy ??

zero_cholesterol = (df['cholesterol'] == 0).sum()
print (f"Total instances of cholesterol with zero value: {zero_cholesterol}")
```

```python
## 172 zero will be a problem to visualize data.
## max cholesterol is 603 and min is 0 , if I take mean of the data it will shift upward , meaning mean will be too high because of few high number
## median will be better choice to replace it
```

```python
## calculating median of only cholesterol columns and replacing with zero

median_cholesterol = df[df['cholesterol']>0]['cholesterol'].median()
df['cholesterol'] = df['cholesterol'].replace(0,median_cholesterol)
df['cholesterol'].describe()
```

Dealing with resting Bp s Column

```python
## Data Set has minimun resting blood pressure 0, which is not possible in a living person

zero_resting_bp = (df['resting bp s']==0).sum()
print(f"Resting Bp with zero value: {zero_resting_bp}")
```

```python
## I have 1 zero value
## max resitng bp is 200 which is medically possible and not a error outlier
## So I can take a mean and replace it because one replaced instance won't make big difference in visualisation unlike cholesterol (172 instances)
```

```python
## replacing zero with mean
mean_resting_bp = df[df['resting bp s']>0]['resting bp s'].mean()
df['resting bp s'] = df['resting bp s'].replace(0, mean_resting_bp)

df['resting bp s'].describe()
```

Dealing with Oldpeak Column

```python
## I also have old peak values in negative
## In dataset those negative values are considered rare clinical variants.
```

```python
## total negative oldpeak value

negative_value_oldpeak = (df['oldpeak']<0).sum()
print(f"Total negative oldpeak records in dataset: {negative_value_oldpeak}")
```

```python
df[df['oldpeak']<0].head(918)
```

```python
## most of the people with negative oldpeak have target 1 ( heart disease positive)
## Negative oldpeaks is pointing towards the problem. So I cannot make it 0
## Best way I can make it positive value
```

```python
df['oldpeak'] = df['oldpeak'].abs()
df['oldpeak'].describe()
```
